### Collecting Current Chart Data

```python
#ChartData(name, date=None, year=None, fetch=True, timeout=25)
def current_search(chart_name): 
    chart_results = billboard.ChartData(chart_name) 
    
    song_title = [] 
    song_artist = []  
    current_rank = [] 
    number_of_weeks_on_chart = [] 
    peak_position = [] 
    last_position = [] 
    song_image = []


    for song in chart_results: 
        song_title.append(song.title) 
        song_artist.append(song.artist) 
        current_rank.append(song.rank)
        number_of_weeks_on_chart.append(song.weeks)
        peak_position.append(song.peakPos) 
        last_position.append(song.lastPos) 
        song_image.append(song.image)
    
    
    
    chart = pd.DataFrame({'artist_name': song_artist,
                          'track_name': song_title, 
                          'rank': current_rank, 
                          'weeks_on_chart' : number_of_weeks_on_chart,
                          'peak_position' : peak_position,
                          'last_position' : last_position, 
                          'song_image': song_image})  
    
    return chart 
```

### Collecting Year-End Chart Data

```python
#ChartData(name, date=None, year=None, fetch=True, timeout=25)
def year_search(chart_name, years): 
    song_title = [] 
    song_artist = []  
    current_rank = [] 
    song_image = []
    chart = []
    year_list = []
    
    for set_year in years: 
        chart_results = billboard.ChartData(chart_name, year=set_year) 
    
        for song in chart_results: 
            song_title.append(song.title) 
            song_artist.append(song.artist) 
            current_rank.append(song.rank) 
            song_image.append(song.image)
            chart.append(chart_name)
            year_list.append(set_year)
    
    
    chart = pd.DataFrame({'artist_name': song_artist,
                          'track_name': song_title, 
                          'rank': current_rank, 
                          'song_image' : song_image,
                          'chart' : chart,
                          'year' : year_list}) 
    
    chart.groupby('year')
    
    return chart 
```

### Collecting Chart Data Over Time

```python
from datetime import datetime

datelist = pd.date_range('2020-01-01', datetime.today()).strftime('%Y-%m-%d').tolist()

datelist
```

```python
def dates_current_search(chart_name, datelist): 
    for i in datelist: 
        chart_results = billboard.ChartData(chart_name, date=i) 
    
        song_title = [] 
        song_artist = []  
        current_rank = [] 
        number_of_weeks_on_chart = [] 
        peak_position = [] 
        last_position = [] 
        song_image = []
        dates = []


        for song in chart_results: 
            song_title.append(song.title) 
            song_artist.append(song.artist) 
            current_rank.append(song.rank)
            number_of_weeks_on_chart.append(song.weeks)
            peak_position.append(song.peakPos) 
            last_position.append(song.lastPos) 
            song_image.append(song.image)
            dates.append(i)
    
    
    
        chart = pd.DataFrame({'artist_name': song_artist,
                              'track_name': song_title, 
                              'rank': current_rank, 
                              'weeks_on_chart' : number_of_weeks_on_chart,
                              'peak_position' : peak_position,
                              'last_position' : last_position, 
                              'song_image': song_image,
                              'dates' : dates})  
    
    return chart 
```

```python
Track_Date_Charts = {}

for name in Track_Charts:
    Track_Date_Charts[name] = dates_current_search(name, datelist)
    
    
Artist_Date_Charts = {}

for name in Artist_Charts:
    Artist_Date_Charts[name] = dates_current_search(name, datelist)
    

Album_Date_Charts = {}

for name in Album_Charts:
    Album_Date_Charts[name] = dates_current_search(name, datelist)
    


```
